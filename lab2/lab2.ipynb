```python
import matplotlib.pyplot as plt
import csv
from lab2 import grad_desc

x_data, y_data = [], []
with open('lr.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        x_data.append(float(row['x'].strip()))
        y_data.append(float(row['y'].strip()))

history = grad_desc(thres=0.01, lr=0.001, x_list=x_data, y_list=y_data)
final_w0, final_w1 = history[-1]

# 3. Simple Plot
plt.figure(figsize=(8, 5))

plt.scatter(x_data, y_data, label='Actual')

line_x = [min(x_data), max(x_data)]
line_y = [final_w1 * x + final_w0 for x in line_x]
plt.plot(line_x, line_y, color='red', label='Model')


plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
