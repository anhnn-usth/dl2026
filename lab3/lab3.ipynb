```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import csv
from lab3 import grad_desc

x1_data, x2_data, y_data = [], [], []

with open('loan2.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        x1_data.append(float(row['Experience'].strip()))
        x2_data.append(float(row['Salary'].strip()))
        y_data.append(float(row['Loan'].strip()))

history = grad_desc(thres=0.1, lr=0.1, x1_list=x1_data, x2_list=x2_data, y_list=y_data)

final_w0, final_w1, final_w2 = history[-1]

plt.figure(figsize=(10, 6))

for i in range(len(y_data)):
    dot_color = 'red' if y_data[i] == 1 else 'blue'
    plt.scatter(x1_data[i], x2_data[i], c=dot_color, edgecolors='black', s=100, alpha=0.8)

line_x = [min(x1_data), max(x1_data)]
line_y = [min(x2_data), max(x2_data)]

plt.plot(line_x, line_y, color='black', linewidth=1, label='Decision Boundary')

plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Logistic Regression: Loan Approval Data & Model Boundary")

from matplotlib.lines import Line2D
custom_legend = [
    Line2D([0], [0], color='black', lw=3, label='Model (0.5 Probability)'),
    Line2D([0], [0], marker='o', color='w', label='Loan Approved (1)', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Loan Denied (0)', markerfacecolor='blue', markersize=10)
]
plt.legend(handles=custom_legend)

plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```
